```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
car_data=pd.read_excel(r"C:\Users\hamza\Downloads\Automobile.xlsx")
print(car_data)
```

```python
data_cleaned=car_data.dropna()
print(data_cleaned)
```

```python
usa_cars=car_data.groupby("origin").get_group("usa")
print("\n",usa_cars)
```

```python
europe_cars=car_data.groupby("origin").get_group("europe")
print("\n",europe_cars)
```

```python
japan_cars=car_data.groupby("origin").get_group("japan")
print("\n",japan_cars)
```

```python
for  origin in ["usa","europe","japan"]:
    print(f"\n{origin} cars:\n", car_data.groupby("origin").get_group(origin))
```

```python
plt.figure(figsize=(15, 10)) 
plt.plot(europe_cars['horsepower'], europe_cars['acceleration'], marker='o', linestyle='-')
plt.xlabel("Horsepower")
plt.ylabel("Acceleration")
plt.title("Horsepower vs Acceleration for European Cars")
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.bar(usa_cars['horsepower'], usa_cars['acceleration'])
plt.xlabel("Horsepower")
plt.ylabel("Acceleration")
plt.title("Horsepower vs Acceleration for USA Cars")
plt.show()
```

```python
plt.figure(figsize=(10,6 )) 
plt.scatter(japan_cars['horsepower'], japan_cars['acceleration'])
plt.xlabel("horsepower")
plt.ylabel("acceleration")
plt.title("graph ")
plt.grid(True)

plt.show()
```

```python
car_counts = data_cleaned['origin'].value_counts()
car_counts = car_counts[car_counts.index.isin(['usa', 'europe', 'japan'])]
ax = car_counts.plot.pie(autopct='%1.1f%%', colors=['blue', 'green', 'red'], figsize=(8, 6))
plt.title("Percentage of Cars by Origin (USA, Europe, Japan)")
plt.ylabel('')  
plt.show()
```

```python

usa_mpg = car_data[car_data['origin'] == 'usa']['mpg']
japan_mpg = car_data[car_data['origin'] == 'japan']['mpg']
europe_mpg = car_data[car_data['origin'] == 'europe']['mpg']

plt.figure(figsize=(10, 6))
plt.boxplot([usa_mpg, japan_mpg, europe_mpg], 
            labels=['USA', 'Japan', 'Europe'])
plt.title('MPG Distribution by Vehicle Origin', fontsize=14)
plt.xlabel('Country of Origin', fontsize=12)
plt.ylabel('Miles Per Gallon (MPG)', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
