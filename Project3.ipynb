```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars_df=pd.read_excel(r"C:\Users\Useer\Downloads\Automobile (2).xlsx")
cars_df
```

```python

# average tempertue of these origin is 
def assign_temperature(origin):
    if origin == 'usa':
        return '70°F'
    elif origin == 'europe':
        return '60°F'
    elif origin == 'japan':
        return '65.5°F'
    else:
        return 'Unknown'
cars_df['temperature'] =cars_df['origin'].apply(assign_temperature)

print(cars_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.boxplot(x=cars_df['temperature'], y=cars_df['displacement'] )
plt.xlabel("Temperature (°F)")
plt.ylabel("Displacement")
plt.title("Box Plot of Displacement by Temperature")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=cars_df['temperature'], y=cars_df['horsepower'])
plt.xlabel("Temperature (°F)")
plt.ylabel("horsepower")
plt.title("Box Plot of Displacement by Temperature")
plt.legend_labels = {'USA': '70°F', 'Japan': '65.5°F', 'Europe': '60°F'}
plt.show()
```
